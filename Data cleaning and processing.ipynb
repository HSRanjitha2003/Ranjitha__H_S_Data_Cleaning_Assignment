```python
import os
import pandas as pd

# Input/Output paths
input_folder = r"Payout Summary & Order Level Sales"
output_file = r"Data Analyst Quick Assignment.xlsx"

summary_data = []

for file in os.listdir(input_folder):
    if file.endswith(".xlsx"):
        file_path = os.path.join(input_folder, file)
        xl = pd.ExcelFile(file_path)

        try:
            df = xl.parse("Summary", header=None)  # Read Summary sheet specifically

            brand = str(df.iloc[4, 1]).strip()
            location = str(df.iloc[5, 1]).strip()
            city = str(df.iloc[6, 1]).strip()
            res_id_text = str(df.iloc[7, 1]).strip()
            res_id = res_id_text.split('-')[-1].strip() if '-' in res_id_text else ""

            payout_period = str(df.iloc[11, 2]).strip()
            payout_settlement_date = str(df.iloc[12, 2]).strip()
            total_payout = str(df.iloc[13, 2]).strip()
            total_orders = str(df.iloc[14, 2]).strip()
            bank_utr = str(df.iloc[15, 2]).strip()

            summary_data.append({
                "Brand": brand,
                "Location": location,
                "City": city,
                "Res-Id": res_id,
                "Payout Period": payout_period,
                "Payout Settlement Date": payout_settlement_date,
                "Total Payout": total_payout,
                "Total Orders (Delivered + Cancelled)": total_orders,
                "Bank UTR": bank_utr,
                "File Name": file
            })

        except Exception as e:
            print(f" Error processing file {file}: {e}")

# Create and save DataFrame to output file
summary_df = pd.DataFrame(summary_data)

with pd.ExcelWriter(output_file, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    summary_df.to_excel(writer, sheet_name="Summary", index=False)

print("Summary tab successfully updated!")
```

```python
import os
import pandas as pd

# Define paths
input_folder = "Payout Summary & Order Level Sales"
output_file = "Data Analyst Quick Assignment.xlsx"

payout_rows = []

# Sort file list to keep order consistent
file_list = sorted([f for f in os.listdir(input_folder) if f.endswith(".xlsx")])

for idx, file in enumerate(file_list, start=1):
    file_path = os.path.join(input_folder, file)

    try:
        # Extract Brand, Res-Id, Payout Period from Summary sheet
        summary_df = pd.read_excel(file_path, sheet_name="Summary", header=None)
        brand = str(summary_df.iloc[4, 1]).strip()
        res_id = summary_df.iloc[7, 1].split('-')[-1].strip()
        payout_period = str(summary_df.iloc[11, 2]).strip()

        # Read Payout Breakup sheet (no headers)
        payout_df = pd.read_excel(file_path, sheet_name="Payout Breakup", header=None)

        # Extract values from fixed cell positions (D5, E5, F5)
        delivered = payout_df.iloc[3, 3]  # Row 4, Col 4
        cancelled = payout_df.iloc[3, 4]  # Row 4, Col 5
        total = payout_df.iloc[3, 5]      # Row 4, Col 6

        # Create one-row DataFrame
        row_data = {
            "SR.No": idx,
            "Delivered Orders": delivered,
            "Cancelled Orders": cancelled,
            "Total": total,
            "Brand": brand,
            "Res-Id": res_id,
            "Payout Period": payout_period,
            "File Name": file
        }

        payout_rows.append(pd.DataFrame([row_data]))

    except Exception as e:
        print(f" Error in {file}: {e}")

# Combine and write
if payout_rows:
    final_df = pd.concat(payout_rows, ignore_index=True)

    with pd.ExcelWriter(output_file, engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
        final_df.to_excel(writer, sheet_name="Payout BreakUp", index=False)

    display(final_df)
else:
    print("No data extracted from any file.")
```

```python
from spire.pdf import PdfDocument, FileFormat
import os

# Specify the directory containing the PDF files
directory = r"Commission Invoices"

# Iterate over all files in the specified directory
for filename in os.listdir(directory):
    if filename.endswith(".pdf"):
        # Construct full file path
        file_path = os.path.join(directory, filename)
        
        # Check if the file exists
        if os.path.exists(file_path):
            # Create an object of the PdfDocument class
            pdf = PdfDocument()
            # Load the PDF file
            pdf.LoadFromFile(file_path)

            # Define output file path
            output_file_path = os.path.join(directory, filename.replace('.pdf', '.xlsx'))

            # Save the PDF file to Excel XLSX format
            pdf.SaveToFile(output_file_path, FileFormat.XLSX)
            # Close the PdfDocument object
            pdf.Close()

            # Optionally, remove the original PDF file
            os.remove(file_path)

            print(f"Converted and replaced: {filename} to {output_file_path}")
        else:
            print(f"File does not exist: {filename}. Please check the file path.")
```

```python
import pandas as pd

# File paths
input_file = 'Commission Invoices/Swiggy_Tax_Sample_file.xlsx'
output_workbook = 'Data Analyst Quick Assignment.xlsx'

# Load Excel
df = pd.read_excel(input_file)

# Columns to extract
required_columns = [
    'id', 'payout_period', 'file_name', 'year', 'month', 'irn', 
    'description', 'quantity', 'unit_price', 'base_amount', 
    'discount', 'assessable_value', 'cgst_amount', 'sgst_amount', 
    'total_amount', 'grand_total'
]

# Rename map
renamed_columns = {
    'id': 'Order ID',
    'payout_period': 'Payout Period',
    'file_name': 'File Name',
    'year': 'Order Year',
    'month': 'Order Month',
    'irn': 'Res-Id',
    'description': 'Order Description',
    'quantity': 'Item Total',
    'unit_price': 'Packaging Charges',
    'base_amount': 'Net Bill Value (before taxes)',
    'discount': 'Restaurant Discounts',
    'assessable_value': 'GST Collected',
    'cgst_amount': 'Total Customer Paid',
    'sgst_amount': 'Commission charged on',
    'total_amount': 'Long Distance Charges',
    'grand_total': 'Net Payout for Order'
}

# Mapping of File Name to Brand & Res-Id
file_brand_mapping = {
    "taco_Tax_Invoice_408_09042025_250409FS29010820": ("Alpha Biryani", "408"),
    "taco_Tax_Invoice_4780_09042025_250409FS29009158": ("Dilli Darbar", "4780"),
    "taco_Tax_Invoice_180796_09042025_250409FS29011371": ("Biryani Box", "180796"),
    "taco_Tax_Invoice_180801_09042025_250409FS29007230": ("Roj Ka Khana(Daily Meals)", "180801"),
    "taco_Tax_Invoice_231392_09042025_250409FS29007041": ("Chinese Chaska", "231392"),
    "taco_Tax_Invoice_313834_09042025_250409FS29011069": ("Biryani Hub", "313834"),
    "taco_Tax_Invoice_330114_09042025_250409FS29011158": ("Ramji Cafe", "330114"),
    "taco_Tax_Invoice_444423_09042025_250409FS29002733": ("Uruvalu Biryani", "444423"),
    "taco_Tax_Invoice_445384_09042025_250409FS29006872": ("Mann Food Court", "445384"),
    "taco_Tax_Invoice_570268_09042025_250409FS29006449": ("Koolerz", "570268"),
    "taco_Tax_Invoice_593472_09042025_250409FS29003032": ("Grill Factory", "593472"),
    "taco_Tax_Invoice_600278_09042025_250409FS29010197": ("Dilli Chaat", "600278")
}

# Filter and rename
filtered_df = df[required_columns].copy()
filtered_df.rename(columns=renamed_columns, inplace=True)

# Add Brand and fixed Res-Id from file_name
filtered_df["Brand"] = filtered_df["File Name"].apply(
    lambda x: file_brand_mapping.get(x.replace(".pdf", "").replace(".xlsx", ""), ("", ""))[0]
)
filtered_df["Res-Id (Mapped)"] = filtered_df["File Name"].apply(
    lambda x: file_brand_mapping.get(x.replace(".pdf", "").replace(".xlsx", ""), ("", ""))[1]
)

# Save to "Order_Level" sheet
with pd.ExcelWriter(output_workbook, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    filtered_df.to_excel(writer, sheet_name='Order_Level', index=False)

print(" Data successfully written to 'Order_Level' tab in", output_workbook)

# Display the final DataFrame
filtered_df
```
